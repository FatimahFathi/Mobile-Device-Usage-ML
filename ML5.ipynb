```python
import matplotlib.pyplot as plt
df.groupby("User_Behavior_Class")["App_Usage_Time"].mean().plot(kind="bar")

plt.title("Average App Usage Time Per Behavior Class")
plt.xlabel("Behavior Class")
plt.ylabel("Average App Usage Time(minutes)")

plt.show()
```
